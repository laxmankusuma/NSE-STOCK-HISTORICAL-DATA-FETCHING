```
#to know which stock is older one
finall['Date'] = pd.to_datetime(finall['Date'])
finall[['SYMBOL','Date']].groupby('SYMBOL').min().reset_index().sort_values('Date', ascending=True)
```

```
finall['year'] = pd.DatetimeIndex(finall['Date']).year
finall['month'] = pd.DatetimeIndex(finall['Date']).month
finall['day'] = pd.DatetimeIndex(finall['Date']).day
finall['week'] = finall['Date'].dt.week
finall['day_name'] = finall['Date'].dt.day_name()
```
